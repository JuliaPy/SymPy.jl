<h1>Gotchas</h1>

<p><a href="https://docs.sympy.org/latest/tutorial/gotchas.html">From</a></p>

<p>To begin, we should make something about SymPy clear.  SymPy is nothing more than a Python library, like <code>NumPy</code>, <code>Django</code>, or even modules in the Python standard library <code>sys</code> or <code>re</code>.  What this means is that SymPy does not add anything to the Python language.  Limitations that are inherent in the Python language are also inherent in SymPy.  It also means that SymPy tries to use Python idioms whenever possible, making programming with SymPy easy for those already familiar with programming with Python.  As a simple example, SymPy uses Python syntax to build expressions.  Implicit multiplication (like <code>3x</code> or <code>3 x</code>) is not allowed in Python, and thus not allowed in SymPy. To multiply <code>3</code> and <code>x</code>, you must type <code>3*x</code> with the <code>*</code>.</p>

<h5>In <code>Julia</code>:</h5>

<ul>
<li><p>implicit multiplication by a <em>literal</em> is supported, unlike Python</p>
</li>
</ul>

<h2>Symbols</h2>

<p>One consequence of this fact is that SymPy can be used in any environment where Python is available.  We just import it, like we would any other library:</p>

```julia
    >>> from sympy import *
```

<h5>In <code>Julia</code>:</h5>

<ul>
<li><p>the functions from the <code>sympy</code> module are loaded with the package:</p>
</li>
</ul>

```julia
using SymPy
```

<hr />

<p>This imports all the functions and classes from SymPy into our interactive Python session.  Now, suppose we start to do a computation.</p>

```julia
    >>> x + 1
    Traceback (most recent call last):
    ...
    NameError: name 'x' is not defined
```

<h5>In <code>Julia</code>:</h5>

<ul>
<li><p>the error output may differ, but an <code>UndefVarError</code> is thrown</p>
</li>
</ul>

```julia
x + 1
```

<hr />

<p>Oops! What happened here?  We tried to use the variable <code>x</code>, but it tells us that <code>x</code> is not defined.  In Python, variables have no meaning until they are defined.  SymPy is no different.  Unlike many symbolic manipulation systems you may have used, in SymPy, variables are not defined automatically. To define variables, we must use <code>symbols</code>.</p>

```julia
    >>> x = symbols('x')
    >>> x + 1
    x + 1
```

<h5>In <code>Julia</code>:</h5>

```julia
x = symbols("x")
x + 1
```

<hr />

<p><code>symbols</code> takes a string of variable names separated by spaces or commas, and creates Symbols out of them.  We can then assign these to variable names. Later, we will investigate some convenient ways we can work around this issue. For now, let us just define the most common variable names, <code>x</code>, <code>y</code>, and <code>z</code>, for use through the rest of this section</p>

```julia
    >>> x, y, z = symbols('x y z')
```

<h5>In <code>Julia</code>:</h5>

```julia
x, y, z = symbols("x y z")
```

<hr />

<p>As a final note, we note that the name of a Symbol and the name of the variable it is assigned to need not have anything to do with one another.</p>

```julia
    >>> a, b = symbols('b a')
    >>> a
    b
    >>> b
    a
```

<h5>In <code>Julia</code>:</h5>

```julia
a, b = symbols("b a")
a
```

```julia
b
```

<hr />

<p>Here we have done the very confusing thing of assigning a Symbol with the name <code>a</code> to the variable <code>b</code>, and a Symbol of the name <code>b</code> to the variable <code>a</code>.  Now the Python variable named <code>a</code> points to the SymPy Symbol named <code>b</code>, and visa versa.  How confusing.  We could have also done something like</p>

```julia
    >>> crazy = symbols('unrelated')
    >>> crazy + 1
    unrelated + 1
```

<h5>In <code>Julia</code>:</h5>

```julia
crazy = symbols("unrelated")
crazy + 1
```

<hr />

<p>This also shows that Symbols can have names longer than one character if we want.</p>

<p>Usually, the best practice is to assign Symbols to Python variables of the same name, although there are exceptions:  Symbol names can contain characters that are not allowed in Python variable names, or may just want to avoid typing long names by assigning Symbols with long names to single letter Python variables.</p>

<p>To avoid confusion, throughout this tutorial, Symbol names and Python variable names will always coincide.  Furthermore, the word "Symbol" will refer to a SymPy Symbol and the word "variable" will refer to a Python variable.</p>

<p>Finally, let us be sure we understand the difference between SymPy Symbols and Python variables.  Consider the following::</p>

```julia
  x = symbols('x')
  expr = x + 1
  x = 2
  print(expr)
```

<p>What do you think the output of this code will be?  If you thought <code>3</code>, you're wrong.  Let's see what really happens</p>

```julia
    >>> x = symbols('x')
    >>> expr = x + 1
    >>> x = 2
    >>> print(expr)
    x + 1
```

<h5>In <code>Julia</code>:</h5>

<ul>
<li><p>we must change to double quotes (or use <code>@vars x</code>, say)</p>
</li>
</ul>

```julia
x = symbols("x")
expr = x + 1
x = 2
expr
```

<hr />

<p>Changing <code>x</code> to <code>2</code> had no effect on <code>expr</code>.  This is because <code>x &#61; 2</code> changes the Python variable <code>x</code> to <code>2</code>, but has no effect on the SymPy Symbol <code>x</code>, which was what we used in creating <code>expr</code>.  When we created <code>expr</code>, the Python variable <code>x</code> was a Symbol.  After we created, it, we changed the Python variable <code>x</code> to 2.  But <code>expr</code> remains the same.  This behavior is not unique to SymPy.  All Python programs work this way: if a variable is changed, expressions that were already created with that variable do not change automatically.  For example</p>

```julia
    >>> x = 'abc'
    >>> expr = x + 'def'
    >>> expr
    'abcdef'
    >>> x = 'ABC'
    >>> expr
    'abcdef'
```

<h5>In <code>Julia</code>:</h5>

<ul>
<li><p>The <code>*</code> infix operator is used for string concatenation</p>
</li>
</ul>

```julia
x = "abc"
expr = x * "def"
expr
```

```julia
x = "ABC"
expr
```

<div class="admonition note"><p class="admonition-title">Quick Tip</p></div>

<p>To change the value of a Symbol in an expression, use <code>subs</code></p>

```julia
     >>> x = symbols('x')
     >>> expr = x + 1
     >>> expr.subs(x, 2)
     3
```

<h5>In <code>Julia</code>:</h5>

```julia
x = symbols("x")
expr = x + 1
expr.subs(x, 2)
```

<p>Or use the "call" form of <code>subs</code>: <code>expr&#40;x &#61;&gt; 2&#41;</code></p>

<hr />

<p>In this example, if we want to know what <code>expr</code> is with the new value of <code>x</code>, we need to reevaluate the code that created <code>expr</code>, namely, <code>expr &#61; x &#43; 1</code>.  This can be complicated if several lines created <code>expr</code>.  One advantage of using a symbolic computation system like SymPy is that we can build a symbolic representation for <code>expr</code>, and then substitute <code>x</code> with values.  The correct way to do this in SymPy is to use <code>subs</code>, which will be discussed in more detail later.</p>

```julia
    >>> x = symbols('x')
    >>> expr = x + 1
    >>> expr.subs(x, 2)
    3
```

<h5>In <code>Julia</code>:</h5>

```julia
x = symbols("x")
expr = x + 1
expr.subs(x, 2)
```

<h2>Equals signs</h2>

<p>Another very important consequence of the fact that SymPy does not extend Python syntax is that <code>&#61;</code> does not represent equality in SymPy.  Rather it is Python variable assignment.  This is hard-coded into the Python language, and SymPy makes no attempts to change that.</p>

<p>You may think, however, that <code>&#61;&#61;</code>, which is used for equality testing in Python, is used for SymPy as equality.  This is not quite correct either.  Let us see what happens when we use <code>&#61;&#61;</code>.</p>

```julia
    >>> x + 1 == 4
    False
```

<h5>In <code>Julia</code>:</h5>

<ul>
<li><p><code>&#61;&#61;</code> is similar as in Python:</p>
</li>
</ul>

```julia
x + 1 == 4
```

<p>Recall <code>&#61;&#61;</code> promotes values, so we have a Julia object may be "equal" to a <code>SymPy</code> one:</p>

```julia
0 == zero(Sym)  ## or Sym(0)
```

<hr />

<p>Instead of treating <code>x &#43; 1 &#61;&#61; 4</code> symbolically, we just got <code>False</code>.  In SymPy, <code>&#61;&#61;</code> represents exact structural equality testing.  This means that <code>a &#61;&#61; b</code> means that we are <em>asking</em> if <code>a &#61; b</code>.  We always get a <code>bool</code> as the result of <code>&#61;&#61;</code>.  There is a separate object, called <code>Eq</code>, which can be used to create symbolic equalities</p>

```julia
    >>> Eq(x + 1, 4)
    Eq(x + 1, 4)
```

<h5>In <code>Julia</code>:</h5>

```julia
Eq(x + 1, 4)
```

<hr />

<p>There is one additional caveat about <code>&#61;&#61;</code> as well.  Suppose we want to know if $(x + 1)^2 = x^2 + 2x + 1$.  We might try something like this:</p>

```julia
    >>> (x + 1)**2 == x**2 + 2*x + 1
    False
```

<h5>In <code>Julia</code>:</h5>

```julia
(x + 1)^2 == x^2 + 2*x + 1
```

<hr />

<p>We got <code>False</code> again. However, $(x + 1)^2$ <em>does</em> equal $x^2 + 2x + 1$. What is going on here?  Did we find a bug in SymPy, or is it just not powerful enough to recognize this basic algebraic fact?</p>

<p>Recall from above that <code>&#61;&#61;</code> represents <em>exact</em> structural equality testing. "Exact" here means that two expressions will compare equal with <code>&#61;&#61;</code> only if they are exactly equal structurally.  Here, $(x + 1)^2$ and $x^2 + 2x + 1$ are not the same symbolically. One is the power of an addition of two terms, and the other is the addition of three terms.</p>

<p>It turns out that when using SymPy as a library, having <code>&#61;&#61;</code> test for exact structural equality is far more useful than having it represent symbolic equality, or having it test for mathematical equality.  However, as a new user, you will probably care more about the latter two.  We have already seen an alternative to representing equalities symbolically, <code>Eq</code>.  To test if two things are equal, it is best to recall the basic fact that if <code>a &#61; b</code>, then <code>a - b &#61; 0</code>.  Thus, the best way to check if <code>a &#61; b</code> is to take <code>a - b</code> and simplify it, and see if it goes to 0.  We will learn :ref:<code>later &lt;tutorial-simplify&gt;</code> that the function to do this is called <code>simplify</code>. This method is not infallible–-in fact, it can be <code>theoretically proven &lt;http://en.wikipedia.org/wiki/Richardson&#37;27s_theorem&gt;</code>_ that it is impossible to determine if two symbolic expressions are identically equal in general–-but for most common expressions, it works quite well.</p>

```julia
    >>> a = (x + 1)**2
    >>> b = x**2 + 2*x + 1
    >>> simplify(a - b)
    0
    >>> c = x**2 - 2*x + 1
    >>> simplify(a - c)
    4*x
```

<h5>In <code>Julia</code>:</h5>

```julia
a = (x + 1)^2
b = x^2 + 2*x + 1
simplify(a - b)
```

```julia
c = x^2 - 2*x + 1
simplify(a - c)
```

<hr />

<p>There is also a method called <code>equals</code> that tests if two expressions are equal by evaluating them numerically at random points.</p>

```julia
    >>> a = cos(x)**2 - sin(x)**2
    >>> b = cos(2*x)
    >>> a.equals(b)
    True
```

<h5>In <code>Julia</code>:</h5>

```julia
a = cos(x)^2 - sin(x)^2
b = cos(2*x)
a.equals(b)
```

<h2>Two Final Notes: <code>^</code> and <code>/</code></h2>

<p>You may have noticed that we have been using <code>**</code> for exponentiation instead of the standard <code>^</code>.  That's because SymPy follows Python's conventions.  In Python, <code>^</code> represents logical exclusive or.  SymPy follows this convention:</p>

```julia
     >>> True ^ False
     True
     >>> True ^ True
     False
     >>> x^y
     Xor(x, y)
```

<h5>In <code>Julia</code>:</h5>

<ul>
<li><p>we export <code>True</code> and <code>False</code> for symbolic Boolean values</p>
</li>
<li><p>This does <strong>not</strong> apply, as we use <code>^</code> for exponentiation.</p>
</li>
<li><p>Use the prefix <code>Or</code> for logical</p>
</li>
</ul>

```julia
Or(True, False)
```

```julia
Or(True, True)
```

```julia
Or(x, y)
```

<hr />

<p>Finally, a small technical discussion on how SymPy works is in order.  When you type something like <code>x &#43; 1</code>, the SymPy Symbol <code>x</code> is added to the Python int <code>1</code>.  Python's operator rules then allow SymPy to tell Python that SymPy objects know how to be added to Python ints, and so <code>1</code> is automatically converted to the SymPy Integer object.</p>

<p>This sort of operator magic happens automatically behind the scenes, and you rarely need to even know that it is happening.  However, there is one exception.  Whenever you combine a SymPy object and a SymPy object, or a SymPy object and a Python object, you get a SymPy object, but whenever you combine two Python objects, SymPy never comes into play, and so you get a Python object.</p>

```julia
    >>> type(Integer(1) + 1)
    <class 'sympy.core.numbers.Integer'>
    >>> type(1 + 1)
    <... 'int'>
```

<h5>In <code>Julia</code>:</h5>

<ul>
<li><p>In Julia, most operations between <code>SymPy</code> objects and <code>Julia</code> objects will promote to a <code>SymPy</code> objects, but of course <code>Julia</code> objects combined will produce <code>Julia</code> Objects:</p>
</li>
</ul>

```julia
typeof(sympy.Integer(1) + 1)
```

```julia
typeof(1 + 1)
```

<p>To convert a <code>Julia</code> object to a <code>SymPy</code> object, the <code>Sym</code> constructor may be useful:</p>

```julia
Sym(1)
```

<p>To convert a <code>SymPy</code> object to a <code>Julia</code> object, the <code>N</code> function is useful for numbers and booleans:</p>

```julia
N(Sym(1)), N(True)
```

<p>And the <code>lambdify</code> function can produce a function from an expression:</p>

```julia
ex = x^2 - 2x + 2
fn = lambdify(ex)
fn(1) - ex(1)
```

<p>On running the example above in SymPy Live, (1+1) is wrapped    by Integer, so it does not show the correct output.</p>

<p>This is usually not a big deal. Python ints work much the same as SymPy Integers, but there is one important exception:  division.  In SymPy, the division of two Integers gives a Rational:</p>

```julia
    >>> Integer(1)/Integer(3)
    1/3
    >>> type(Integer(1)/Integer(3))
    <class 'sympy.core.numbers.Rational'>
```

<h5>In <code>Julia</code>:</h5>

```julia
sympy.Integer(1)/sympy.Integer(3)
```

```julia
typeof(sympy.Integer(1)/sympy.Integer(3))
```

<p>And to get the Python, type, we can use <code>__class__</code>:</p>

```julia
(sympy.Integer(1)/sympy.Integer(3)).__class__
```

<hr />

<p>But in Python <code>/</code> represents either integer division or floating point division, depending on whether you are in Python 2 or Python 3, and depending on whether or not you have run <code>from __future__ import division</code>:</p>

```julia
    >>> from __future__ import division
    >>> 1/2 #doctest: +SKIP
    0.5
```

<h5>In <code>Julia</code>:</h5>

<ul>
<li><p>This does not apply, as <code>/</code> is not integer division.</p>
</li>
</ul>

<hr />

<p>!! note</p>

<p>On running the example above in SymPy Live, (1/2) is wrapped    by Integer, so it does not show the correct output.</p>

<p>To avoid this, we can construct the rational object explicitly</p>

```julia
    >>> Rational(1, 2)
    1/2
```

<h5>In <code>Julia</code>:</h5>

<ul>
<li><p><code>Rational</code> from <code>sympy</code> is <em>not</em> exported, it would conflict with <code>Julia</code>'s <code>Rational</code> costructor. We must qualify it:</p>
</li>
</ul>

```julia
Rational(1, 2)
```

```julia
sympy.Rational(1, 2)
```

<hr />

<p>This problem also comes up whenever we have a larger symbolic expression with <code>int/int</code> in it.  For example:</p>

```julia
    >>> x + 1/2 #doctest: +SKIP
    x + 0.5
```

<h5>In <code>Julia</code>:</h5>

<ul>
<li><p><code>Int/Int</code> will produce a floating point value, whereas <code>Int//Int</code> will produce a rational, which can then be promoted without loss to a symbolic object:</p>
</li>
</ul>

```julia
x + 1/2 #doctest: +SKIP
```

<hr />

<div class="admonition note"><p class="admonition-title">Note</p></div>

<p>On running the example above in SymPy Live, (1/2) is wrapped    by Integer, so it does not show the correct output.</p>

<p>This happens because Python first evaluates <code>1/2</code> into <code>0.5</code>, and then that is cast into a SymPy type when it is added to <code>x</code>.  Again, we can get around this by explicitly creating a Rational:</p>

```julia
    >>> x + Rational(1, 2)
    x + 1/2
```

<h5>In <code>Julia</code>:</h5>

```julia
x + 1//2
```
